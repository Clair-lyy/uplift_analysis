* 特征说明：
* X:
* Recency: Months since last purchase. 距离上次购买时间
* History_Segment: Categorization of dollars spent in the past year. 过去一年消费金额（分类变量）
* History: Actual dollar value spent in the past year.过去一年消费金额（连续变量）
* Mens: 1/0 indicator, 1 = customer purchased Mens merchandise in the past year.过去一年是否购买男性商品
* Womens: 1/0 indicator, 1 = customer purchased Womens merchandise in the past year.过去一年是否购买女性商品
* Zip_Code: Classifies zip code as Urban, Suburban, or Rural.城市、郊区或农村
* Newbie: 1/0 indicator, 1 = New customer in the past twelve months.是否过去12月内的新用户
* Channel: Describes the channels the customer purchased from in the past year.用户购买渠道

* Treat:
* Mens E-Mail 男性版电子邮件营销活动
* Womens E-Mail 女性版电子邮件营销活动
* No E-Mail 无电子邮件营销活动

* Y:
* Visit: 1/0 indicator, 1 = Customer visited website in the following two weeks. 用户在接下来两周是否浏览网站
* Conversion: 1/0 indicator, 1 = Customer purchased merchandise in the following two weeks.用户在接下来两周是否购买商品
* Spend: Actual dollars spent in the following two weeks.用户在接下来两周的消费金额
    
* 问题：数据集中缺少conversion和spend标签，原始数据中有，dataset中没有，用原始数据

* 提出假设
* 假设1: 电子邮件营销活动对不同用户接下来两周是否浏览网站的影响存在异质性
* 假设2: 电子邮件营销活动对不同用户接下来两周是否购买商品的影响存在异质性
* 假设3: 电子邮件营销活动对不同用户在接下来两周消费金额的影响存在异质性

* 分别讨论男性版电子营销活动、女性版电子营销活动对用户接下来两周是否浏览网站、接下来两周是否购买商品、接下来两周消费金额的个体异质性处理效应，并提炼有明显提升效果的人群共性

## 第二部分：数据预处理/特征工程

```python
### 将有序变量处理成有序数字
### history_segment代表过去一年消费金额分档，是有序变量
set(data['history_segment'])
```

```python
def parse_history_segment(history_segment):
    if history_segment == '1) $0 - $100':
        return 1
    if history_segment == '2) $100 - $200':
        return 2
    if history_segment == '3) $200 - $350':
        return 3
    if history_segment == '4) $350 - $500':
        return 4
    if history_segment == '5) $500 - $750':
        return 5
    if history_segment == '6) $750 - $1,000':
        return 6
    if history_segment == '7) $1,000 +':
        return 7
    return -1
```

```python
df = data.copy()
```

```python
df['history_segment'] = data['history_segment'].apply(parse_history_segment)
```

```python
set(df['history_segment'])
```

```python
### 城市、郊区、乡村是有序变量，也可以尝试处理成哑变量，对比效果
set(data['zip_code'])
```

```python
def parse_zip_code(zip_code):
    if zip_code == 'Rural':
        return 1
    if zip_code == 'Surburban':
        return 2
    if zip_code == 'Urban':
        return 3
```

```python
df['zip_code'] = data['zip_code'].apply(parse_zip_code)
```

```python
set(df['zip_code'])
```

```python
### 将无序变量处理成one-hot
### channel用户来源渠道是无序变量
set(data['channel'])
```

```python
def parse_channel_phone(channel):
    if channel != 'Web':
        return 1
    return 0

def parse_channel_web(channel):
    if channel != 'Phone':
        return 1
    return 0
```

```python
df['channel_phone'] = data['channel'].apply(parse_channel_phone)
df['channel_web'] = data['channel'].apply(parse_channel_web)
```

```python
# 在多维下的条件平均处理效应，均值、CATE
def agg_cate(data_df):
    print(data_df)
    label = 'visit'
    mens = data_df[data_df['segment_mens']==1]
    womens = data_df[data_df['segment_womens']==1]
    
    base = data_df[(data_df['segment_mens']==0) & (data_df['segment_womens']==0)]
    
    base_mean = base[label].mean()
    mens_mean = mens[label].mean()
    womens_mean = womens[label].mean()
    
    mens_uplift = mens_mean-base_mean
    womens_uplift = womens_mean-base_mean
    
    return base_mean, mens_mean, womens_mean, mens_uplift, womens_uplift
```

```python
def get_cuts_mean(df, cuts):
    cuts_mean = []
    for i in range(0, len(cuts)):
        if i == 0:
            mean_value = df[df['uplift'] > cuts[i]]['uplift'].mean()
        else:
            mean_value = df[(df['uplift'] <= cuts[i-1]) & (df['uplift'] > cuts[i])]['uplift'].mean()
        cuts_mean.append(mean_value)
    mean_value = df[df['uplift'] <= cuts[-1]]['uplift'].mean()
    cuts_mean.append(mean_value)
    return cuts_mean

def get_uplift_cut(x, cuts):
    percent = scipy.stats.percentileofscore(cuts, x)/100
    index = int(percent*(len(cuts)+1)) if int(percent*(len(cuts)+1)) <= len(cuts) else len(cuts)
    return len(cuts)-index

def plot_uplift(df, bins=10, treatment='treatment', metric='label', is_plot=True):
    cuts = [np.percentile(df['uplift'], i) for i in range(100-int(100/bins),0 , -int(100/bins))]
    cuts_mean = get_cuts_mean(df,cuts)
    df['pre_uplift_score'] = df['uplift'].apply(lambda x: cuts_mean[get_uplift_cut(x, cuts)])
    df2 = df.groupby([treatment, 'pre_uplift_score'])[metric].mean().reset_index()
    act_df = pd.DataFrame()
    act_df['uplift_score'] = df2[df2[treatment]==1].reset_index()[metric] - df2[df2[treatment]==0].reset_index()[metric]
    act_df['type'] = 'actual'
    pre_df = pd.DataFrame()
    pre_df['uplift_score'] = df2[df2[treatment]==0]['pre_uplift_score']
    pre_df['type'] = 'predict'
    new_df = act_df.append(pre_df)
    if is_plot:
        print(new_df)
        sns.set(rc={'figure.figsize':(11.7,8.27)})
        #sns.barplot(x=new_df.index, y='uplift_score', hue='type', data=new_df, order=[i for i in range(bins-1,-1,-1)])
        sns.barplot(x=new_df.index, y='uplift_score', hue='type', data=new_df, order=[i for i in range(bins-1,-1,-1)])
    return new_df
```

```python
X_train, X_test = train_test_split(df, test_size=0.3, random_state=1)
X_train, X_verify = train_test_split(X_train, test_size=0.3, random_state=1)
#X_columns = ['recency', 'history_segment', 'history', 'mens', 'womens', 'zip_code',
#       'newbie', 'channel_phone', 'channel_web']
#X_columns = ['recency_norm', 'history_segment_norm', 'history_norm', 'mens', 'womens', 'zip_code_norm',
#       'newbie', 'channel_phone', 'channel_web']
X_columns = ['recency', 'history_segment', 'history_norm', 'mens', 'womens', 'zip_code',
       'newbie', 'channel_phone', 'channel_web']
y_columns = 'visit'
# y_columns = 'spend_norm'
```

```python
'''
causal_forest = CausalForestDML(criterion='het', 
                                n_estimators=1000,       
                                min_samples_leaf=100, 
                                max_depth=5, 
                                max_samples=0.5,
                                discrete_treatment=False,
                                honest=True,
                                inference=True,
                                cv=10,
                                model_t=LassoCV(), 
                                model_y=LassoCV(),
                                )
'''
def estimate_model(model, X_data, X_columns, y_columns, treatment):
    # 计算MSE
    mse_score = model.score(X_data[y_columns], X_data[[treatment]], X=X_data[X_columns])
    
    X_data['uplift'] = model.effect(X_data[X_columns])
    # 计算GINI曲线面积
    auuc_info = auuc_score(X_data[[y_columns, treatment, 'uplift']], outcome_col=y_columns, treatment_col=treatment)
    auuc = abs(auuc_info['uplift'] - auuc_info['Random'])
    
    #plot_gain(X_data[[y_columns, treatment, 'uplift']], outcome_col=y_columns, treatment_col=treatment)
    
    # 计算单调性
    new_df = plot_uplift(X_data, treatment=treatment, metric=y_columns, is_plot=False)
    monotonicity = spearmanr(new_df[new_df['type']=='actual'].index, \
                             new_df[new_df['type']=='actual']['uplift_score'])[0]
    
    return mse_score, auuc, monotonicity
    

def train_causal_forest_model(criterion, n_estimators, min_samples_leaf, max_depth, \
                              X_train, X_columns, y_columns, treatment):
    causal_forest = CausalForestDML(criterion=criterion, 
                                n_estimators=n_estimators,  
                                min_samples_leaf=min_samples_leaf, 
                                max_depth=max_depth, 
                                max_samples=0.5,
                                discrete_treatment=False,
                                honest=True,
                                inference=True,
                                cv=3)
    causal_forest.fit(X_train[y_columns], X_train[[treatment]], X=X_train[X_columns], W=None, cache_values=True)
    
    return causal_forest
    
## 按照累积效应寻优
def search_params_by_auuc(params, X_train, X_verify, X_columns, y_columns, treatment='treatment'):
    best_auuc = 0
    best_model = None
    for criterion in params['criterion']:
        for n_estimators in params['n_estimators']:
            for min_samples_leaf in params['min_samples_leaf']:
                for max_depth in params['max_depth']:
                    print('criterion: %s, n_estimators: %s, min_samples_leaf: %s, max_depth: %s' % \
                         (criterion, n_estimators, min_samples_leaf, max_depth)) 
                    start_time = time.time()
                    model = train_causal_forest_model(criterion, n_estimators, min_samples_leaf, \
                        max_depth, X_train, X_columns, y_columns, treatment)
                    end_time = time.time()
                    print('cost time: %ss' % (end_time - start_time))
                    mse_score, auuc, monotonicity = estimate_model(model, X_verify, X_columns, \
                                                                         y_columns, treatment)
                    print('mse_score: %s, auuc: %s, monotonicity: %s' % (mse_score, auuc, monotonicity))
                    
                    if best_auuc < auuc:
                        best_auuc = auuc
                        best_model = model   
    return best_auuc, best_model

## 按照偏序关系寻优
def search_params_by_monotonicity(params, X_train, X_verify, X_columns, y_columns, treatment='treatment'):
    best_monotonicity = 0
    best_model = None
    for criterion in params['criterion']:
        for n_estimators in params['n_estimators']:
            for min_samples_leaf in params['min_samples_leaf']:
                for max_depth in params['max_depth']:
                    print('criterion: %s, n_estimators: %s, min_samples_leaf: %s, max_depth: %s' % \
                         (criterion, n_estimators, min_samples_leaf, max_depth)) 
                    start_time = time.time()
                    model = train_causal_forest_model(criterion, n_estimators, min_samples_leaf, \
                        max_depth, X_train, X_columns, y_columns, treatment)
                    end_time = time.time()
                    print('cost time: %ss' % (end_time - start_time))
                    mse_score, auuc, monotonicity = estimate_model(model, X_verify, X_columns, \
                                                                         y_columns, treatment)
                    print('mse_score: %s, auuc: %s, monotonicity: %s' % (mse_score, auuc, monotonicity))
                    
                    if best_monotonicity < monotonicity:
                        best_monotonicity = monotonicity
                        best_model = model   
    return best_monotonicity, best_model

## 按照累积效应+偏序关系寻优
def search_params(params, X_train, X_verify, X_columns, y_columns, treatment='treatment'):
    best_auuc = 0
    best_model = None
    for criterion in params['criterion']:
        for n_estimators in params['n_estimators']:
            for min_samples_leaf in params['min_samples_leaf']:
                for max_depth in params['max_depth']:
                    print('criterion: %s, n_estimators: %s, min_samples_leaf: %s, max_depth: %s' % \
                         (criterion, n_estimators, min_samples_leaf, max_depth)) 
                    start_time = time.time()
                    model = train_causal_forest_model(criterion, n_estimators, min_samples_leaf, \
                        max_depth, X_train, X_columns, y_columns, treatment)
                    end_time = time.time()
                    print('cost time: %ss' % (end_time - start_time))
                    mse_score, auuc, monotonicity = estimate_model(model, X_verify, X_columns, \
                                                                         y_columns, treatment)
                    print('mse_score: %s, auuc: %s, monotonicity: %s' % (mse_score, auuc, monotonicity))
                    
                    if monotonicity >= 0.5 and best_auuc < auuc:
                        best_auuc = auuc
                        best_model = model   
    return best_auuc, best_model
```

```python
def cut_source(arrs, x, bins, cuts):
    percent = int(scipy.stats.percentileofscore(arrs, x)/bins)
    percent = len(cuts) - 1 if percent == len(cuts) else percent
    return round(cuts[percent],5)

def plot_uplift_source(df, bins=10, treatment='treatment', metric='label'):
    cuts = [scipy.stats.scoreatpercentile(df['uplift'], i) for i in range(0, 100, int(100/bins))]
    df['uplift_score'] = df['uplift'].apply(lambda x: cut_source(df['uplift'], x, bins, cuts))
    df2 = df.groupby([treatment, 'uplift_score'])[metric].mean().reset_index()
    new_df = pd.DataFrame()
    new_df[metric] = df2[df2[treatment]==1].reset_index()[metric] - df2[df2[treatment]==0].reset_index()[metric]
    new_df['uplift_score'] = df2[df2[treatment]==0]['uplift_score']
    #new_df = new_df.reset_index().rename({metric:'%s_rate' % metric},axis=1)
    sns.set(rc={'figure.figsize':(11.7,8.27)})
    sns.barplot(x="uplift_score", y=metric, data=new_df, order=new_df.sort_values(['uplift_score'],ascending=False)['uplift_score'])
    
```
